# Finding the values and generating the graph of Stochastic Oscillator using pandas 


```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Making Function to calculate the Stochastic Oscillator %k and %D
def stochastic_oscillator(data, period=14):
    highs = data['High']
    lows = data['Low']
    closes = data['Close']
 
    #rolling  works on a particular peroid/subset in a dataset  
    highs_max = highs.rolling(window=period).max()
    lows_min = lows.rolling(window=period).min()
 
    #finfing the value of k , k = A line which shows an asset's low and high price 
    #in a specific frame to create a price range , shows last closing price as a percent in the range
    k = 100 * ((closes - lows_min) / (highs_max - lows_min))
    print(k)
 
    #finding the value of d,d = which is average of 3 values of k 
    D = k.rolling(window=3).mean()
    print(D)
    return k,D
 
```

```python
# Read data from CSV
# Dataset taken from the kaggle and it is the dataset provided by the tesla   
file_path = r"C:\Users\hp\Downloads\tesla-stock-price.csv"  
data = pd.read_csv(file_path) 

# Calling the function to calculate the K and D 
k , D  = stochastic_oscillator(data)
```

```python
# Using matlpotlib plotting the results of k and d 
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price')
plt.plot(k, label='%K', color='b')
plt.plot(D, label='%D', color='r')
plt.title('Stochastic Oscillator')
plt.xlabel('Data Points')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
